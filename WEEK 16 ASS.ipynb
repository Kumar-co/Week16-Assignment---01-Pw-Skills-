Q1. What are the three measures of central tendency?

ANS- The three measures of central tendency are:

Mean: The average of a set of values, calculated by adding all the values together and dividing by the number of values.

Median: The middle value in a sorted list of numbers. If there’s an even number of values, the median is the average of the two middle values.

Mode: The value that appears most frequently in a data set. There can be more than one mode if multiple values have the same highest frequency.




Q2. What is the difference between the mean, median, and mode? How are they used to measure the
central tendency of a dataset?

ANS-The mean, median, and mode are all measures of central tendency, but they each capture different aspects of a dataset:

     Mean Definition- The mean is calculated by summing all the values and dividing by the number of values.
     Usage- It provides a single value that represents the overall average of the dataset. However, it can be
     influenced by extreme values (outliers), which can skew the result.

    Median Definition- The median is the middle value when the data is arranged in ascending order. 
    If there's an even number of values, it’s the average of the two middle values.
    Usage- The median is particularly useful when dealing with skewed distributions or outliers, as it better 
    represents the central point of a dataset without being affected by extreme values.

    Mode Definition- The mode is the value that appears most frequently in a dataset.
    Usage- It is helpful in understanding the most common value in the data, especially for categorical data or
    when analyzing frequency distributions. A dataset can have one mode, more than one mode (bimodal or multimodal), or no mode at all.

Q8. What do you understand about skewness in data?

ANS - Skewness refers to the measure of asymmetry in the distribution of data. It helps describe how the values in a dataset are distributed relative to the mean. Skewness can 
         indicate whether the data is symmetric or has a longer tail on one side.

                    Types of Skewness
                      Positive Skewness (Right Skewed):

                          The tail on the right side of the distribution is longer or fatter.
                            The majority of data points are concentrated on the left, with a few extreme values on the right.
                           Mean > Median > Mode.
                             Negative Skewness (Left Skewed):

                                  The tail on the left side of the distribution is longer or fatter.
                                     The majority of data points are concentrated on the right, with a few extreme values on the left.

Q9. If a data is right skewed then what will be the position of median with respect to mean?

ANS- In a right-skewed distribution (positive skewness), the mean is typically greater than the median.
       Relationship:
         Mean > Median

         In a right-skewed distribution, the longer tail on the right side pulls the mean toward the higher values, while the median, being the middle value, remains less affected by the extreme values in the tail. As a result, the mean is shifted to the right of the median.

           Visualization:

              Right Skewed Distribution
                Mean: → (to the right)
                Median: ↓ (to the left)

Q12. For a normal distribution data what is the relationship between its measure of central tendency?

In a normal distribution, the measures of central tendency—namely the mean, median, and mode—are all equal. This characteristic is a defining feature of the normal distribution. Here’s a breakdown:

1. **Mean**: The arithmetic average of all data points.
2. **Median**: The middle value when the data points are arranged in ascending order.
3. **Mode**: The value that appears most frequently in the dataset.

### Relationship in a Normal Distribution
- **Mean = Median = Mode**: In a perfectly normal distribution, these three measures are located at the center of the distribution and have the same value. This symmetry around the mean results in a bell-shaped curve.

This relationship indicates that the normal distribution is symmetric, with half of the data points falling below the mean and half above it.
